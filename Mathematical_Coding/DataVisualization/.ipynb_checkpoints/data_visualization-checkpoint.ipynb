### Problem 1

The file `anscombe.npy` contains Anscombe’s quartet.
For each section of the quartet,
- Plot the data as a scatter plot on the box [0, 20] × [0, 13].
- Use `scipy.stats.linregress()` to calculate the slope and intercept of the least squares regression line for the data and its correlation coefficient (the first three return values).
- Plot the least squares regression line over the scatter plot on the domain $x\in[0,20]$.
- Report the mean and variance in $x$ and $y$, the slope and intercept of the regression line, and the correlation coefficient. Compare these statistics to those of the other sections.
- Describe how the section is similar to the others and how it is different.

```python
A = np.load("anscombe.npy")
for i in range(4):
    plt.scatter(A[:,int(2*i)],A[:,int(2*i)+1],label='Scatter Points')                 #Scatter the points
    slope, intercept, rval = stats.linregress(A[:,int(2*i)],A[:,int(2*i)+1])[:3]      #Get the slop, intercept, and rvalue
    n = np.arange(21)
    plt.plot(n,intercept + slope*n,label='Regression Model')                          #Plot and label each plot
    plt.xlim([0,20])
    plt.ylim([0,13])                                                                  #Set the X and Y limits
    plt.legend(loc='upper left')
    plt.title('Quartet '+str(int(i+1)))
    plt.show()
    print('X Mean:\t\t',np.mean(A[:,i]))                                              #Print the X and Y means, slop intercept and rvalue
    print('Y Mean:\t\t',np.mean(A[:,i+1]))
    print('Slope:\t\t',slope)
    print('Intercept:\t',intercept)
    print('Correlation Coefficient:',rval)
```

The first and third quartets are different from the second and fourth as they very much resemble linear functions. The second quartet resembles a quadratic function and and the fourth looks like a vertical function. I find it very interesting that they all have the same slopes and y-intercepts.  I think this shows that a linear regression line may not be very effective for data that does not resemble a linear function, and for data that has a lot of noise. Saying that, I find it very intersting that they all have around the same correlation coefficient. I think that these numbers are very important to look at, but looking at the visualization of the actual plot is also very important for accurately describing the data. 

### Problem 2
The $n + 1$ Bernstein basis polynomials of degree $n$ are defined as follows.

$$b_{n,v}(x) = \binom{n}{v}x^v(1-x)^{n-v},\quad v=0,\ 1,\ \ldots,\ n$$

Plot the first $10$ Bernstein basis polynomials ($n = 0,\ 1,\ 2,\ 3$) as small multiples on the domain $[0, 1] \times [0, 1]$.
Label the subplots for clarity, adjust tick marks and labels for simplicity, and set the window limits of
each plot to be the same.
Consider arranging the subplots so that the rows correspond with $n$ and the columns with $v$.

(Hint: The constant $\binom{n}{v} = \frac{n!}{v!(n-v)!}$ is called the binomial coefficient and can be efficiently computed with `scipy.special.binom()` or `scipy.misc.comb()`.)

```python
x = np.linspace(0,1,100)                                                                   #Create a domain on [0,1]
for n in range(4):
    fig, axes = plt.subplots(1,n+1)                                                        #Create subplots for each n
    for v in range(n+1):
        if n == 0:
            plt.plot(x, scipy.special.binom(n,v)*(x**v)*(1-x)**(n-v))                      #Use the plot function if n=0
            plt.axis([0,1,0,2])
            plt.title('n = '+str(n) + ' v = ' + str(v))
        else:
            axes[v].plot(x, scipy.special.binom(n,v)*(x**v)*(1-x)**(n-v))                  #If n>0 use the axes functions
            axes[v].axis([0,1,0,2])
            axes[v].set_title('n = '+str(n) + ' v = ' + str(v))                            #Set the titles
    plt.tight_layout()                                                                     #Use a tight layout so none of the figure titles or labels overlap
    plt.show()

```

### Problem 3

The file `MLB.npy` contains measurements from over 1,000 recent Major League Baseball players, compiled by UCLA.
Each row in the array represents a different player; the columns are the player’s height (in inches), weight (in pounds), and age (in years), in that order.

Create several visualizations to show the correlations between height, weight, and age in the MLB data set.
Use at least one scatter plot.
Adjust the marker size, plot a regression line, change the window limits, and use small multiples where appropriate.

```python
height, weight, age = np.load("MLB.npy").T

plt.scatter(height, weight)                                                         #Scatter the height by weight data
slope, intercept, r_value, p_value, std_err = stats.linregress(height, weight)         #Get the data from a linear regression
plt.plot(height, slope*height+intercept, color = 'r')
plt.xlabel("Height in Inches")                                                      #Label the axes and create a title
plt.ylabel("Weight in Pounds")
plt.title("Height By Weight Comparison")
plt.show()
```

```python
plt.bar(height, age)                                              #Create a bar graph for height by 
plt.xlabel("Height in Inches")                                    #Label the axis and title
plt.ylabel("Age in Years")
plt.title('Height by Age Comparison')
plt.show()
```

### Problem 4

The file `earthquakes.npy` contains data from over 17,000 earthquakes between 2000 and 2010 that were at least a 5 on the Richter scale.
Each row in the array represents a different earthquake; the columns are the earthquake’s date (as a fraction of the year), magnitude (on the Richter scale), longitude, and latitude, in that order.

Describe the data with two or three visualizations, including line plots, scatter plots, and histograms as appropriate.
Your plots should answer the following questions:
1. How many earthquakes happened every year?
2. How often do stronger earthquakes happen compared to weaker ones?
3. Where do earthquakes happen? Where do the strongest earthquakes happen?

Hint: Use `plt.axis("equal")` or `ax.set_aspect("equal")` to fix the aspect ratio, which may improve comparisons between longitude and latitude.

```python
years, magnitudes, longitude, latitude = np.load("earthquakes.npy").T

plt.hist(years)                                                          #Plot a histogram of the number of earthquakes per year
plt.xlabel("Years")                                                      #Label axis and create a title
plt.ylabel("Number of Earthquakes")
plt.title("Number of Earthquakes per Year")                              #Answers question how many earthquakes happen per year
plt.show()
```

```python
plt.hist(magnitudes,bins=50)                                      #Histogram of earthquakes at a certain magnitude
plt.ylabel("Number of Earthquakes")                               #Label axis
plt.xlabel("Magnitudes of Earthquakes")                           #Answers the question, how often stronger earthquakes happen compared to weaker ones
plt.title('Frequency of Earthquakes by Magnitude')
plt.show()
```

```python
mask = magnitudes>=7                                                             #Create a mask for stronger earthquakes
plt.scatter(longitude, latitude,s=3, label="All Earthquakes")                    #Scatter all of the earthquakes by their latitudes and longitudes
plt.scatter(longitude[mask],latitude[mask],s=3, label="Earthquakes Above 7")     #Scatter all earthquakes above a magnitude of 7
plt.legend()
plt.xlabel("Longitude")                                                          #Label the axis and create a title
plt.ylabel("Latitude")
plt.axis("equal")                                                                #Answers queston where do earthquakes occur, and where are the stronger ones.
plt.title("Geographic Location of Earthquakes")
plt.show()
```

### Problem 5

The _Rosenbrock function_ is defined as follows. $$f(x,y)=(1−x)^2 +100(y−x^2)^2$$
The minimum value of $f$ is $0$, which occurs at the point $(1, 1)$ at the bottom of a steep, banana-shaped valley of the function.

Use a heat map and a contour plot to visualize the Rosenbrock function.
Plot the minimizer $(1, 1)$.
Use a different sequential colormap for each visualization.

```python
x = np.linspace(-1,2,200)          
fig, axes = plt.subplots(1,2)
X,Y = np.meshgrid(x,x)                                                                        #Create a domain for the Rosenbrock function
Z = (1-X)**2+100*(Y-X**2)**2
axes[0].pcolormesh(X,Y,Z,cmap='viridis')                                                      #Create a heat map
axes[0].set_title("Heat Map of Rosenbrock Function")
axes[1].contourf(X,Y,Z,[-.5,-.25,0,.5,.75,1,1.25], cmap='magma')                              #Create a contour map that shows the banana-shaped valley
axes[1].set_title("Contour Map of Rosenbrock Function")
fig.colorbar(axes[0].pcolormesh(X,Y,Z,cmap='viridis'), ax=axes[0])
fig.colorbar(axes[1].contourf(X,Y,Z,[-.5,-.25,0,.5,.75,1,1.25], cmap='magma'), ax=axes[1])    #Insert color bars
axes[1].scatter(1,1, color='b',s=10, label='Minimizer')                                       #Plot the minimizer (1,1,0)
axes[1].legend()
fig.tight_layout()
plt.show()
```

The file `countries.npy` contains information from 20 different countries.
Each row in the array represents a different country; the columns are the 2015 population (in millions of people), the 2015 GDP (in billions of US dollars), the average male height (in centimeters), and the average female height (in centimeters), in that order.

Visualize this data set with at least four plots, using at least one scatter plot, one histogram, and one bar chart.
List the major insights that your visualizations reveal.

(Hint: consider using `np.argsort()` and fancy indexing to sort the data for the bar chart.)

```python
countries = np.array(
            ["Austria", "Bolivia", "Brazil", "China", "Finland",
             "Germany", "Hungary", "India", "Japan", "North Korea",
             "Montenegro", "Norway", "Peru", "South Korea", "Sri Lanka",
             "Switzerland", "Turkey", "United Kingdom", "United States", "Vietnam"])
population, gdp, male, female = np.load("countries.npy").T

```

```python
plt.barh(countries, population)                                           #Create a horizontal bar graph of countries by population
plt.xlabel("Population in Millions of People")                            #Label the axis and create a title
plt.ylabel("Countries")
plt.title('Population by Country Bar Graph')
plt.tight_layout()
plt.show()
```

```python
plt.hist(male)                                                   #Create a histogram of the male average height per country
plt.title('Average Country Male Height Histogram')               #Label axis and create the title
plt.xlabel('Average Country Height')
plt.ylabel('Number of Countries')
plt.show()
```

```python
plt.scatter(countries, male-female)                                      #Create a scatter plot of the differencesin height of males and females per country
plt.xticks(countries, rotation='vertical')                               #Label the x and y axis and create a title
plt.ylabel('Difference in Height by Centimeters')
plt.title("Difference in Male and Female Height by Country")
plt.show()
```

```python
plt.scatter(countries, gdp)                                    #Scatter plot the GDP by country
plt.xticks(countries, rotation='vertical')                     #Turn the country names sideways
plt.title('2015 GDP in U.S. Dollars')                          #Lable the axis and create a title
plt.ylabel('Worth in Billions of U.S. Dollars')
plt.show()
```
